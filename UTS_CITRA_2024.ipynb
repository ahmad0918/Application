# Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Baca gambar


```python
img = cv2.imread('gambar_nama.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

# Range warna

```python
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])
lower_red_alt = np.array([170, 50, 50])
upper_red_alt = np.array([180, 255, 255])

lower_green = np.array([40, 50, 50])
upper_green = np.array([80, 255, 255])
```

# Deteksi warna

```python
mask_blue = cv2.inRange(hsv_img, lower_blue, upper_blue)
mask_red = cv2.inRange(hsv_img, lower_red, upper_red) + cv2.inRange(hsv_img, lower_red_alt, upper_red_alt)
mask_green = cv2.inRange(hsv_img, lower_green, upper_green)
```

# Tampilkan gambar dan hasil deteksi warna

```python
plt.figure(figsize=(30, 8))
plt.subplot(1, 4, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(30, 8))
plt.subplot(1, 4, 2)
plt.imshow(mask_blue, cmap='gray')
plt.title('Deteksi Warna Blue')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(30, 8))
plt.subplot(1, 4, 3)
plt.imshow(mask_red, cmap='gray')
plt.title('Deteksi Warna Red')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(30, 8))
plt.subplot(1, 4, 4)
plt.imshow(mask_green, cmap='gray')
plt.title('Deteksi Warna Green')
plt.axis('off')
plt.show()
```
